Se creará un DataFrame el cual tendrá como columnas: IDUsuario, Fecha, Llave, Calificacion. Cada fila del dataframe representa una retroalimentación por parte del usuario cuando se realiza una recomendación. El usuario unicamente puede dicidir si le gustó o no la recomendación a través de botones de like y dislike por lo cual el atributo calificación solo tendrá dos valores. Si a el usuario le agrada la recomendación se da un valor de 1 si no -1. Esto con el fin de modificar los pesos que tienen los usuarios por dewey cuando se generan los clusters y en caso de que las recomendaciones no sean adecuadas cambiar al usuario de cluster. 

```python
calificaciones = pd.DataFrame(columns= ["IDUsuario", "Year", "IDItem", "Calificación"])
calificaciones
```

```python
import random
def aleatorios(num):
    for i in range(num):
        id = random.choice(df.IDUsuario.unique())
        IDItem = random.choice(df.IDItem.unique())
        cal = random.choice([-1,1])
        anio = random.choice([2021])
        calificaciones.loc[i] = [id,anio,IDItem,cal]
```

```python
aleatorios(100)
```

```python
calificaciones
```

```python
def aleatorioCluster(num, cluster):
    sub = df.loc[df.cluster == cluster]
    for i in range(num):
        id = random.choice(sub.IDUsuario.unique())
        IDItem = random.choice(df.IDItem.unique())
        cal = random.choice([-1,1])
        anio = random.choice([2021])
        calificaciones.loc[i] = [id,anio,IDItem,cal]
```

```python
aleatorioCluster(100,3)
```

```python
def librosRelacionados(num, cluster):
    sub = df.loc[df.cluster == cluster]
    for i in range(num):
        id = random.choice(sub.IDUsuario.unique())
        IDItem = random.choice(sub.IDItem.unique())
        cal = random.choice([-1,1])
        anio = random.choice([2021])
        calificaciones.loc[i] = [id,anio,IDItem,cal]
```

```python
aleatorioCluster(100,3)
```

```python
def agregarUsuario(id, anio, IDItem, cal):
    i = calificaciones.shape[0]
    calificaciones.loc[i] = [id,anio,IDItem,cal]
```
